# 1. Introdução

As séries de Fourier permitem a análise de sinais periódicos. Um sinal periódico possui repetições de valores dentro de um determinado intervalo definido como período:

$ f(t) = f(t + kT) $, em que $T$ é o período.

**Exemplo** seno e cosseno:

<p style="text-align:center">$ cos(0) = cos(2\pi) = cos(4\pi) = \dots = cos(2k\pi) $</p>
<p style="text-align:center">$ T = 2\pi $</p>

# 2. Espaço Lp

Lembrando da norma euclidiana:

$ \| x \|_{2} = \left( \sum |x_{i}|^{2} \right)^{1/2} $, também chamada de norma $l^2$

$ \| x \|_{p} = \left( \sum |x_{i}|^{p} \right)^{1/p} $, também chamada de norma $l^p$

Definimos a norma $L_{p}$ como:

<p style="text-align:center">$ \| f \|_{p} = \left( \displaystyle \int_D |f(x)|^{p} dx \right)^{1/p} $ </p>

Espaço $L_2$: é o espaço das funções quadrado integráveis

<p style="text-align:center">$ L_{2} = \left\{ f: \Re \Rightarrow \mathbb{C} , \left( \int | f(x) |^2 dx \right)^{1/2} < \infty \right\} $  </p>



Seja $V$ o espaço das funções $ V = \left\{ g:[ -\pi,\pi ] \implies \Re, ( \displaystyle \int_{-\pi}^{\pi} |g(x)|^2 dx )^{1/2} < \infty \right\} $

Vamos definir os seguintes conceitos:

<p style="text-align:center">a) $ \langle p,q \rangle = \displaystyle \int_{-\pi}^{\pi} p(x) q(x) dx $ </p>

<p style="text-align:center">b) $ \| p \|^{2} = \displaystyle \int_{-\pi}^{\pi} |p(x)|^2 dx$ </p>

Escolhendo um subespaço de $V$ chamado $S$, tal que a base de $S$ seja $ S:\left\{ 1, cos(x), sen(x), cos(2x), sen(2x), \dots, cos(nx), sen(nx) \right\} $

A dimensão de $S$ depende de $n$: $dim(S) = 2n+1$

Qualquer elemento $g_n \in S$ é uma combinação linear dos elementos da base.

Ou seja:

<p style="text-align:center">$ g_n \in S: g_n(x) = \displaystyle \sum_{n=0}^{N} a_n cos(nx) + \displaystyle \sum_{n=1}^{N} b_n sen(x) $</p>

Se $g_n$ é a projeção ortogonal de uma função $g \in V$ sobre $S$, então $a_n$ e $b_n$ são os **Coeficientes de Fourier**.

<p style="text-align:center">$ a_n = \dfrac{\langle g, cos(nx) \rangle}{\langle cos(nx),cos(nx) \rangle}, n=0,1,\dots,N $</p>

<p style="text-align:center">$ b_n = \dfrac{\langle g, sen(nx) \rangle}{\langle sen(nx),sen(nx) \rangle}, n=1,2,\dots,N $</p>

Relembrando dos conceitos anteriores (a) e (b), podemos entender melhor os coeficientes:

$ \langle g, cos(nx) \rangle = \displaystyle \int_{-\pi}^{\pi} g(x) cos(nx) dx $

$ \langle cos(nx),cos(nx) \rangle = \displaystyle \int_{-\pi}^{\pi} cos^2(nx) dx = \pi $

<p style="text-align:center">$ a_n = \dfrac{\langle g, cos(nx) \rangle}{\langle cos(nx),cos(nx) \rangle} = \dfrac{1}{\pi} \displaystyle \int_{-\pi}^{\pi} g(x) cos(nx) dx , n=1,\dots,N $</p>

<p style="text-align:center">$ b_n = \dfrac{\langle g, sen(nx) \rangle}{\langle sen(nx),sen(nx) \rangle} = \dfrac{1}{\pi} \displaystyle \int_{-\pi}^{\pi} g(x) sen(nx) dx , n=1,\dots,N $</p>

<p style="text-align:center">$ a_0 = \dfrac{\langle g, 1 \rangle}{\langle 1,1 \rangle} = \dfrac{\int_{-\pi}^{\pi} g(x)dx}{\int_{-\pi}^{\pi} 1 dx} = \dfrac{1}{2\pi} \displaystyle \int_{-\pi}^{\pi} g(x) dx $</p>

Assim, podemos definir a **Soma Parcial da Série de Fourier**:

<p style="text-align:center">$ g_{N(x)} = a_0 + \displaystyle \sum_{n=1}^{N} a_n cos(nx) + \displaystyle \sum_{n=1}^{N} b_n sen(nx)  $</p>

# 3. Exemplos

Considere a função $f:[-1,1] =
  \begin{cases}
    0       & \quad \text{se } -1 \leq t \leq 0\\
    1  & \quad \text{se } 0 < t \leq 1
  \end{cases}
$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.arange(-1,1,0.001)
f = t.copy()  #t -> tempo

for count,i in enumerate(t):
    if (i >= -1) and (i <= 0):
        f[count]=0
    elif (i > 0) and (i <= 1):
        f[count]=1

# cont = índice e enumerate é uma função do for que contou do período o espaçamento de 0,001;

plt.plot(t,f)
plt.show()
```

Vamos calcular os coeficientes de Fourier para a função $f(t)$

$ a_0 = \dfrac{1}{2} \displaystyle \int_{-1}^{1} f(t) dt = \dfrac{1}{2} \displaystyle \int_{0}^{1} 1 dt = \dfrac{1}{2} $

$ a_0 = \dfrac{1}{2} $

;

$ a_n = \dfrac{1}{2} \displaystyle \int_{-1}^{1} f(t) cos\left(\dfrac{2\pi n t}{2}\right) dt $

$ = \dfrac{1}{2}  \displaystyle \int_{0}^{1} 1 \cdot cos(\pi n t) dt =  \left. \dfrac{1}{2} \dfrac{sen(\pi n t)}{\pi n} \right|_0^1 $

$ = \dfrac{1}{2} \dfrac{sen(\pi n t)}{\pi n} = \dfrac{1}{2} sinc(t) $

$a_n =
  \begin{cases}
    1/2       & \quad \text{se } n = 0\\
    0  & \quad \text{se } n \neq 0
  \end{cases}
$

;

$ b_n = \dfrac{1}{2} \displaystyle \int_{-1}^{1} f(t) sen\left(\dfrac{2\pi n t}{2}\right) dt $

$ = \dfrac{1}{2} \displaystyle \int_{0}^{1} 1 \cdot sen(\pi n t) dt = \dfrac{1}{2} \left( \left. -\left[ -\dfrac{cos(\pi n t)}{\pi n} \right]  \right|_0^1 \right) $

$ = \dfrac{1}{2} \left( \dfrac{cos(\pi n)}{\pi n} - \dfrac{cos(0)}{\pi n} \right) $

$b_n =
  \begin{cases}
    0,       & \quad \text{se } n \text{ par}\\
    1/\pi n,  & \quad \text{se } n \text{ ímpar}
  \end{cases}
$

Assim, a projeção de $f(t)$ no subespaço gerado por $\left\{1, cos(n \pi t), sen(n \pi t), \dots, cos(N \pi t), sen(N \pi t)\right\}$ é dada por

$ f_{N(t)} = \dfrac{1}{2} + \displaystyle \sum \dfrac{2}{\pi(2n-1)} sen(\pi(2n-1)t) $

```python
#Para n=1:

f_0 = 1/2
f_1 = f_0 + (2/(np.pi))*np.sin(np.pi*t)

plt.plot(t,f_1)
plt.show()
```

```python
#Para n=3:

f_3 = f_2 + (2/(5*np.pi))*np.sin(5*np.pi*t)

plt.plot(t,f_3)
plt.show()
```

```python
f_4 = f_3 + (2/(7*np.pi))*np.sin(7*np.pi*t)
f_5 = f_4 + (2/(9*np.pi))*np.sin(9*np.pi*t)
f_6 = f_5 + (2/(11*np.pi))*np.sin(11*np.pi*t)

f_s = [f_1, f_2, f_3, f_4, f_5, f_6]

plt.figure(figsize=(10,6))
plt.title("Soma da Série de Fourier para a função Step")

plt.plot(t,f,"k",alpha=1)

for count,i in enumerate(f_s):
    plt.plot(t,i)
    
plt.show()
```

### Extra

Utilizando o módulo ``scipy``, é possível calcular os coeficientes de fourier diretamente no código. Vamos usar uma função triângulo como exemplo:

```python
import numpy as np
from scipy.signal import square,sawtooth,triang
import matplotlib.pyplot as plt
from scipy.integrate import simps
```

```python
L=2 # Periodicity of the periodic function f(x)
samples=2001

x=np.linspace(0,L,samples,endpoint=False)
y=triang(samples)

L=4
x=np.linspace(0,L,2*samples,endpoint=False)
y=np.concatenate((y,y),axis=0)
```

```python
plt.plot(x,y)
plt.show()
```

```python
# Fourier Coefficients
a0=2./L*simps(y,x)
an=lambda n:2.0/L*simps(y*np.cos(2.*np.pi*n*x/L),x)
bn=lambda n:2.0/L*simps(y*np.sin(2.*np.pi*n*x/L),x)

print(a0)
```

```python
# Series sum

s_s = []

for k in range(15):
    if k == 0:
        s = a0/2
    else:
        s=s+sum([an(k)*np.cos(2.*np.pi*k*x/L)+bn(k)*np.sin(2.*np.pi*k*x/L)])
        s_s.append(s)
        
s_s = np.delete(s_s,0,axis=0) 
```

```python
plt.figure(figsize=(10,6))
plt.title("Soma da Série de Fourier para a função Triângulo")
plt.plot(x,y,"k")

for i in range(len(s_s)):
    plt.plot(x,s_s[i])

plt.show()
```

É possível fazer o mesmo com outras funções do ``scipy``, como a quadrada (``square``) e a "dente de serra" (``sawtooth``). Fique à vontade para modificar o código e experimentar! Há mais opções de ondas na [documentação da biblioteca](https://docs.scipy.org/doc/scipy-0.14.0/reference/signal.html).

```python
L=2
samples=2001
freq=4 
dutycycle=0.5

x=np.linspace(0,L,samples,endpoint=False)
y=square(2.0*np.pi*x*freq/L,duty=dutycycle)

# Fourier Coefficients
a0=2./L*simps(y,x)
an=lambda n:2.0/L*simps(y*np.cos(2.*np.pi*n*x/L),x)
bn=lambda n:2.0/L*simps(y*np.sin(2.*np.pi*n*x/L),x)

s_s = []

for k in range(25):
    if k == 0:
        s = a0/2
    else:
        s=s+sum([an(k)*np.cos(2.*np.pi*k*x/L)+bn(k)*np.sin(2.*np.pi*k*x/L)])
        s_s.append(s)
        
s_s = np.delete(s_s,0,axis=0) 

plt.figure(figsize=(10,6))
plt.title("Soma da Série de Fourier para a função Quadrada")
plt.plot(x,y,"k")

for i in range(len(s_s)):
    plt.plot(x,s_s[i])

plt.show()
```
